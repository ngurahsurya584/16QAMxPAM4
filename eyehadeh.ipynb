```python
import numpy as np
import matplotlib.pyplot as plt

# Define the PAM4 signal parameters
num_bits = 100000
symbol_rate = 10e9
bit_rate = symbol_rate / 2
pulse_width = 1 / (2 * bit_rate)
amplitude = 1

# Generate the random data for the PAM4 signal
data = np.random.randint(0, 4, num_bits)

# Convert the data to PAM4 signal
pam4_signal = []
for bit in data:
    if bit == 0:
        pam4_signal += [-amplitude, -amplitude]
    elif bit == 1:
        pam4_signal += [-amplitude, amplitude]
    elif bit == 2:
        pam4_signal += [amplitude, -amplitude]
    else:
        pam4_signal += [amplitude, amplitude]

# Add Gaussian noise to the signal
noise_amplitude = 0.5
noise = noise_amplitude * np.random.randn(len(pam4_signal))
pam4_signal_with_noise = pam4_signal + noise

# Plot the PAM4 signal with noise
plt.figure(figsize=(10, 5))
plt.plot(pam4_signal_with_noise, linewidth=0.5)
plt.title("PAM4 Signal with Noise")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```

```python
# Plot the eye diagram of the PAM4 signal with noise
plt.figure(figsize=(10, 5))
num_samples_per_symbol = int(pulse_width * symbol_rate)
for i in range(4):
    start = i * num_samples_per_symbol
    end = start + num_samples_per_symbol
    plt.plot(pam4_signal_with_noise[start:end], linewidth=0.5)
plt.title("Eye Diagram of PAM4 Signal with Noise")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.xlim(0, num_samples_per_symbol)
plt.show()
```
